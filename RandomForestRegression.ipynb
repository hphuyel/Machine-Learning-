# Random Forest Regression

### Imort libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Import dataset

```python
dataset = pd.read_csv("Salaries.csv")
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

### Train Random Forest Regression Model on the whole dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0) # put number of trees, that is estimators
regressor.fit(X, y)
```

### Predicting New Result

```python
regressor.predict([[6.5]])
```

### Visualizing Random Forest Regression Result (High Resolution)

```python
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
