# Ploynomial Regression

## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as  np
```

## Importing Dataset

```python
dataset = pd.read_csv("Salaries.csv")
X = dataset.iloc[:, 1:-1].values # leave first column
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

## Training Liner Regression Model on the Whole Dataset

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y) # no train_test here, taking data directly
```

## Visulaizing the Linear Regression Results

```python
plt.scatter(X, y, color="red")
plt.plot(X, lin_reg.predict(X), color="blue")
plt.title("Truth or Bluff (Liner Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

## Training Ploynomial Regression Model on the Whole Dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

## Visualizing the Polynomial Regression Results

```python
plt.scatter(X, y, color="red")
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color="blue")
plt.title("Truth or Bulff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

## Visualizing Ploynomial Regression Results(for higher resolution and smoothe curve)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

## Predicting a new Result with Linear Regression

```python
lin_reg.predict([[6.5]])
```

## Predicting a new Result with Ploynomial Regression

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
